```python
# Importing Libraries that are relevent for the project
import pandas as pd             #for DataSet reading 
import matplotlib.pyplot as plt #for Visualization
import numpy as np
import math as m
```

```python
df=pd.read_csv("SortedDataset.csv")   #Reading the CSV file for DataSet
print(df.head())                #Verify if Import was successful
```

```python
# setting ( x_value, y_value, dotSize, Color)
def displayGraph(lxlimit,hxlimit,ylimit,darkness,title,boolean,xline=[0,0],yline=[0,0]):
    plt.scatter(df['Wind Speed (m/s)'],df['LV ActivePower (kW)'], s=darkness, c='black') 
    plt.xlabel('Wind Speed (m/s)')       # Label for X axis
    plt.ylabel('Active Power (kW)')      # Label for Y axis
    plt.title(title)                     # Title of the scatterplot
    plt.grid(boolean)
    if(boolean):
        plt.plot(xline,yline,color='red', linewidth=2)
        val=np.arange(5.0,12.5,0.5)
        plt.xticks(val)
    plt.xlim(lxlimit,hxlimit)
    plt.ylim(-250,ylimit)
    plt.show()

displayGraph(0,25,3750,0.002,'Scatter plot of dataset',False)
```

```python
displayGraph(6,12,3750,0.009,'Range of Wind speed with Rectilinear Characteristics',False)
```

```python
def CalcSumError(x, y, w:float, b:float):
    sumW=sumB=0.0
    ŷ=(w*x)+b
    error=ŷ-y
    sumW=np.sum(error*x)
    sumB=np.sum(error)
    return (sumW,sumB)

def GradientDescent(x,y):
    w=0.0
    b=0.0
    alpha=0.0001
    tolerance=0.000001
    preverror=[-1,-1]
    currerror=[-2,-2]
    n=len(x)
    while(abs(preverror[0]-currerror[0])>tolerance):
        preverror = currerror
        currerror = CalcSumError(x,y,w,b)
        w=w-(alpha*currerror[0]*(n**-1))
        b=b-(alpha*currerror[1]*(n**-1))
    return (w,b)

def LinearReg():
    fdf=df[(df['Wind Speed (m/s)'] >=6.0) & (df['Wind Speed (m/s)'] <=12.0)]
    x = fdf['Wind Speed (m/s)'].to_numpy()
    y = fdf['LV ActivePower (kW)'].to_numpy()
    n=len(x)
    WandB = GradientDescent(x,y)
    w=WandB[0]
    b=WandB[1]
    xline = [6.0, 12.0]
    yline = [(w*6.0)+b, (w*12.0)+b]
    # print("Error %: ",#???) 
    displayGraph(6,12,3750,0.009,'Model At Work',True,xline,yline)
    
    return (w,b)

def main():
    global l1
    l1=LinearReg()
main()
```
